```python
sales =[1200,450,980,1500,3000]
with open("sales_data.txt","w") as s:
    for sale in sales:
        s.write(str(sale)+"\n")
with open ("sales_data.txt","r") as r:
    print(r.read())
```

```python
with open("sales_data.txt", "r") as r:
    print("--- Full Content ---")
    print(r.read()) 
    r.seek(0) 
    
    # 3. Now readline() will work again
    line1 = r.readline()
    print(f"first line is : {line1.strip()}")
    line2 = r.readline()
    print(f"second line is : {line2.strip()}")

# Your second block is already correct!
with open("sales_data.txt", "r") as r:
    lines = r.readlines()
    print("--- List of Lines ---")
    print(lines)
```

```python
with open("sales_data.txt", "a") as ap:
    ap.write("5000\n2500\n1700\n")
with open("sales_data.txt", "r") as r:
    print("--- Updated Content ---")
    print(r.read())
```

```python
with open("sales_data.txt", "r") as r:
    a = r.read()
    b = a.strip()
    total_sales = sum(map(int, b.split()))
    print(f"Total Sales: {total_sales}")
    highest_sale = max(map(int, b.split()))
    print(f"Highest Sale: {highest_sale}")
    lowest_sale = min(map(int, b.split()))
    print(f"Lowest Sale: {lowest_sale}")
    avg_sale = total_sales / len(b.split())
    print(f"Average Sale: {avg_sale:.2f}")
```
