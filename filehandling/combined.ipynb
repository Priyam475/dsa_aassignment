```python
price = {
    "Mouse": 500,
    "Keyboard": 1500,
    "Monitor": 7000,
    "Pendrive": 300,
    "Camera": 5000
}

# Opening the file once outside the loop is more efficient
with open("discounted_price.txt", "w") as s:  # Using "w" to start fresh each run
    for key, value in price.items():
        # Using an f-string makes the input prompt cleaner
        A = input(f"Enter discount for {key} (Current Price: {value}): ")
        
        # Calculation
        discount = (value * int(A)) / 100
        new_price = value - discount
        
        # Writing the line
        s.write(f"{key} : {value} : {new_price}\n")

# Reading it back
print("\n--- Final Discounted Prices ---")
with open("discounted_price.txt", "r") as r:
    print(r.read())
```
