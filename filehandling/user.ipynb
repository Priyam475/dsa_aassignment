```python
import os 
for i in range(3): 
    filename = input("Enter file name (e.g., sales_data.txt): ")
    

    if os.path.isfile(filename):
        with open(filename, "r") as r:
            print(f"--- Content of {filename} ---")
            print(r.read())
            break  
    else:
        print(f"Error: '{filename}' does not exist. Try again.")
```
