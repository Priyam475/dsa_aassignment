```python
def add_price(prices_list, price):
    """Append a new price to the list and return it."""
    prices_list.append(price)
    return prices_list


def avg_price(prices_list):
    return sum(prices_list) / len(prices_list)


def max_price(prices_list):
    return max(prices_list)


# sample usage with a menu-driven loop
prices = []  # start with empty list

menu = """
1) Add price (append a new price to list)
2) Calculate average price
3) Find maximum price
q) Quit
"""

while True:
    choice = input(menu + "Choose option: ")
    if choice == "1":
        try:
            p = float(input("Enter price to add: "))
        except ValueError:
            print("Please enter a valid number")
            continue
        prices = add_price(prices, p)
        print("Updated prices:", prices)
    elif choice == "2":
        if prices:
            print("Average price:", avg_price(prices))
        else:
            print("No prices available yet")
    elif choice == "3":
        if prices:
            print("Maximum price:", max_price(prices))
        else:
            print("No prices available yet")
    elif choice.lower() == "q":
        print("Exiting menu")
        break
    else:
        print("Invalid selection, try again")
```
