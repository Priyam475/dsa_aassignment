```python
def apply_discount(price, discounted_price):
    if discounted_price == 0:
        d= price*(5/100)
        return price - d 
        
    else :
        e= price*(discounted_price/100)
        return price -e 
    
```

```python
A = apply_discount(1000,13)
print(A)
```

```python
def factorial(n):
    if n<0 :
        print("error can't use negative")
        return None
    elif n==0 or n==1:
        return 1
    return n*factorial(n-1)
print(factorial(5))
```
