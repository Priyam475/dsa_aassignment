```python
A= int(input("Enter the number: "))
B= int(input("Enter the divisor: "))
try :
    C= A/B
    print("The result is: ", C)
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
except ValueError:
    print("Error: Invalid input. Please enter numeric values.")
finally:
    print("Program execution completed.")
```

```python
prices = [120, 350, 'abc', 500, -200, 800]
total = 0

for price in prices:
    try:
        # Check for negative value first to raise custom error
        if isinstance(price, (int, float)) and price < 0:
            raise ValueError("Negative price not allowed")
        
        # This will naturally raise a TypeError if 'price' is a string
        total += price
        print(f"Added {price}. Running total: {total}")

    except TypeError:
        print(f"Skipping '{price}': Not a number.")
    except ValueError as e:
        print(f"Skipping {price}: {e}")

print(f"\nFinal Bill Total: {total}")
```

```python
def check_age(age):
    if age < 1 or age > 120:
        raise ValueError("Age must be between 1 and 120")
    return age

try:
    user_input = input("Enter your age: ")
    # Convert input to integer
    age_val = int(user_input)
    
    # Validate age range
    check_age(age_val)
    print(f"Age {age_val} is valid.")

except ValueError as e:
    # This catches both non-integers and our custom range error
    print(f"Invalid input: {e}")
```

```python
filename = input("Enter the filename to read ( test.txt): ")

try:
    with open(filename, 'r') as file:
        # Read all lines into a list
        lines = file.readlines()
        
        print(f"--- First 3 lines of {filename} ---")
        # Use slicing to get the first 3 lines
        for line in lines[:3]:
            print(line.strip())

except FileNotFoundError:
    print(f"Error: The file '{filename}' was not found.")
except PermissionError:
    print(f"Error: You do not have permission to read '{filename}'.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")

finally:
    # This block executes regardless of success or failure
    print("\nFile reading process complete.")
```

```python
cart = []  

print("Welcome to the Safe Shopping Cart!")
print("Enter item prices or 'q' to finish.")

while True:  # Runs a loop asking user to enter prices
    user_input = input("Enter price: ").lower()
    
    if user_input == 'q':  # Stops when user enters 'q'
        break
        
    try:
        # Convert input to a float
        price = float(user_input)
        
        # Raise custom exception if price is negative
        if price < 0:
            raise ValueError("Price cannot be negative.")
            
        cart.append(price)
        print(f"Added ${price:.2f} to cart.")

    except ValueError as e:
        # Handle ValueError if user enters invalid input
        # This catches both non-numeric text and our custom negative price error
        if "could not convert string to float" in str(e):
            print("Invalid input: Please enter a numeric price.")
        else:
            print(f"Error: {e}")

# At the end, print results
print("-" * 30)
print(f"Total items: {len(cart)}")  # Print Total items
print(f"Total bill: ${sum(cart):.2f}")  # Print Total bill
```
