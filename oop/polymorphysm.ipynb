```python
class Product:
    def __init__(self, name, price, category):
        self.name = name
        self.category = category
        # Task 2: Private attribute
        self.__price = price

    # Task 2: Getter and Setter
    def get_price(self):
        return self.__price

    def set_price(self, new_price):
        if new_price > 0:
            self.__price = new_price
        else:
            print("Invalid price update!")

    def get_info(self):
        print(f"Generic Product: {self.name}")

# Task 4: Inheritance and Polymorphism
class Laptop(Product):
    def get_info(self):
        # Overriding with a unique style
        print(f"💻 [LAPTOP] Model: {self.name} | Price: ${self.get_price()}")

class Mobile(Product):
    def get_info(self):
        # Overriding with a unique style
        print(f"📱 [MOBILE] Device: {self.name} | Category: {self.category}")
```

```python
# Creating a list of different objects
inventory = [
    Laptop("MacBook Pro", 2000, "Electronics"),
    Mobile("iPhone 15", 999, "Phones"),
    Laptop("Dell XPS", 1500, "Electronics")
]

# Task 4: Iterate and show polymorphism
print("--- Inventory Report ---")
for item in inventory:
    item.get_info()
```
