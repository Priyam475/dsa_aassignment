```python
from abc import ABC, abstractmethod

class Payment(ABC):
    @abstractmethod
    def process_payment(self, amount):
        pass

class CreditCardPayment(Payment):
    def process_payment(self, amount):
        print(f"💳 Processing Credit Card payment of ${amount}...")

class UPIPayment(Payment):
    def process_payment(self, amount):
        print(f"📲 Processing UPI payment of ${amount} via QR scan...")
```

```python
class Product:
    def __init__(self, name, price, category):
        self.name = name
        self.category = category
        self.__price = price # Task 2: Private attribute

    def get_price(self):
        return self.__price

    # Task 6.1: Readable string representation
    def __str__(self):
        return f"Product(name={self.name}, price={self.__price}, category={self.category})"

    # Task 6.2: Operator Overloading for addition
    def __add__(self, other):
        return self.__price + other.get_price()

# Subclasses from Task 4
class Laptop(Product):
    def get_info(self):
        print(f"Laptop: {self.name} - ${self.get_price()}")

class Mobile(Product):
    def get_info(self):
        print(f"Mobile: {self.name} - {self.category}")
```

```python
# 1. Test Magic Methods (Task 6)
p1 = Product("Laptop", 1000, "Tech")
p2 = Product("Mouse", 50, "Accessories")

print(p1)  # Uses __str__
total = p1 + p2  # Uses __add__
print(f"Combined Price: ${total}")

# 2. Test Abstraction (Task 5)
cc = CreditCardPayment()
upi = UPIPayment()

cc.process_payment(total)
upi.process_payment(total)
```
