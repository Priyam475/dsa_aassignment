```python
class Product:
    def __init__(self, name, price, category):
        self.name = name
        self.category = category
        self.__price = price # Private attribute (Task 2)

    def get_price(self):
        return self.__price

    def __str__(self): # Magic Method (Task 6)
        return f"Product(name={self.name}, price={self.__price}, category={self.category})"

    def __add__(self, other): # Operator Overloading (Task 6)
        return self.__price + other.get_price()

class Inventory:
    def __init__(self):
        self.products = [] # List to store product objects

    def add_product(self, product):
        self.products.append(product)

    def remove_product(self, name):
        self.products = [p for p in self.products if p.name != name]

    def get_total_value(self):
        return sum(p.get_price() for p in self.products)

    def show_all_products(self):
        for p in self.products:
            print(p)

class Store:
    def __init__(self, store_name):
        self.store_name = store_name
        self.inventory = Inventory() # Composition: Store has an Inventory

    def add_new_product(self):
        # Taking input to create a Product object
        name = input("Enter product name: ")
        price = float(input("Enter price: "))
        category = input("Enter category: ")
        new_prod = Product(name, price, category)
        self.inventory.add_product(new_prod)

    def show_summary(self):
        print(f"\n--- {self.store_name} Summary ---")
        print(f"Total Items: {len(self.inventory.products)}")
        print(f"Total Inventory Value: ${self.inventory.get_total_value()}")
```

```python
# 1. Creating a Store object
my_store = Store("Tech Haven")

# 2. Adding 3 products (without using input)
products_data = [
    ("Laptop", 999.99, "Electronics"),
    ("Mouse", 29.99, "Accessories"),
    ("Keyboard", 79.99, "Accessories")
]

for name, price, category in products_data:
    product = Product(name, price, category)
    my_store.inventory.add_product(product)

# 3. Showing summary
my_store.show_summary()
my_store.inventory.show_all_products()

# 4. Using __add__ to combine prices of two products
if len(my_store.inventory.products) >= 2:
    p1 = my_store.inventory.products[0]
    p2 = my_store.inventory.products[1]
    combined_price = p1 + p2
    print(f"\nCombined price of {p1.name} and {p2.name}: ${combined_price}")
```
