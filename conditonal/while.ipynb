```python
orders = []

while True:
    print("\n--- Menu ---")
    print("1 — Add order amount")
    print("2 — Show all orders and totals (with discounts)")
    print("q — Quit")
    
    choice = input("Choose an action: ").lower()

    if choice == 'q':
        print("Exiting program...")
        break  # Stops the loop entirely
    
    if choice == '1':
        amount = float(input("Enter order amount: "))
        orders.append(amount)
        print("Order added.")
    
    elif choice == '2':
        print("\nOrder Summary:")
        for price in orders:
            # Applying your discount logic from earlier
            if price >= 2000:
                final = price * 0.85
            elif price >= 1500:
                final = price * 0.90
            elif price >= 1000:
                final = price * 0.93
            else:
                final = price
            print(f"Original: {price} | Discounted: {final:.2f}")
            
    else:
        print("Invalid choice, please try again.")
        continue  
```
