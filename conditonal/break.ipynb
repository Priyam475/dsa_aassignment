```python
daily = [200, 150, 0, 400, 50, -1, 300]
total_sales = 0

for sale in daily:
    if sale == -1:
        print("Corrupted data found (-1). Stopping process.")
        break  # Immediately exits the for loop
    
    if sale == 0:
        print("No sales today, skipping...")
        continue  # Skips the addition and moves to the next item in the list
    
    # This part only runs if the checks above don't trigger break/continue
    total_sales += sale
    print(f"Added {sale}. Running total: {total_sales}")

print("-" * 20)
print(f"Final Total Sales: {total_sales}")
```
