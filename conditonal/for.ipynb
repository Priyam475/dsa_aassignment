```python

prices = [1200, 2500, 800, 1750,3000]

print("Batch Discount Report:")
print("-" * 25)

for p in prices:
    if p >= 2000:
        discounted = p * 0.85
    elif p >= 1500:
        discounted = p * 0.90
    elif p >= 1000:
        discounted = p * 0.93
    else:
        discounted = p
        
    print(f"Original: {p} -> After Discount: {discounted:.2f}")
```
