```python
"""
TASK 1: Load International Education Costs Dataset

Purpose:
    Load the main dataset from a CSV file for analysis.
    This dataset contains information about education costs across different
    countries, universities, and programs.

Dataset: International_Education_Costs.csv
Expected columns: Country, University, Program, Level, Duration_Years, etc.
"""

import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("International_Education_Costs.csv")

print("Dataset loaded successfully!")
print(f"Shape: {df.shape}")
```

```python
"""
Explore Dataset Structure and Create Derived Features

Operations:
    1. Display dataset shape and dimensions
    2. Create 'total_cost' feature by combining cost components
    3. Verify column names

Note: Total cost is computed as the sum of all cost-related columns.
This gives a comprehensive measure of the total education expenses.
"""

# Display dataset dimensions
print("\n=== Dataset Shape ===")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")

# Create total cost feature using snake_case naming
if 'Living_Cost_Index' in df.columns:
    df['total_cost_index'] = (
        df['Living_Cost_Index'] + 
        df['Visa_Fee_USD'] + 
        df['Rent_USD'] + 
        df['Insurance_USD']
    )

# Display column names
print("\n=== Column Names ===")
print(df.columns.tolist())
```

```python
"""
Data Quality Check: Missing Values and Data Preview

Purpose:
    Identify missing values in each column and display sample data.
    Missing data handling is crucial for reliable analysis.

Output:
    - Count of null values per column
    - Percentage of missing data
    - First 5 rows of the dataset
"""

# Check for missing values
print("\n=== Missing Values Analysis ===")
missing_values = df.isnull().sum()
if missing_values.sum() == 0:
    print("✓ No missing values found!")
else:
    print("Missing values by column:")
    missing_percentage = (missing_values / len(df)) * 100
    for col, count in missing_values[missing_values > 0].items():
        print(f"  {col}: {count} ({missing_percentage[col]:.2f}%)")

# Display first 5 rows
print("\n=== Data Preview ===")
df.head()
```

```python
"""
Select and Prepare Key Features for Analysis

Purpose:
    Select relevant columns for the main analysis.
    These columns provide the core information for EDA:
    - Geographic: Country
    - Institution: University
    - Academic: Program, Level, Duration
    - Economic: Exchange_Rate, Total_cost

Rationale:
    This selection focuses on essential features that directly relate to
    education costs while eliminating redundant or unnecessary columns.
"""

# Define columns of interest with consistent naming
columns_of_interest = [
    'Country',
    'University',
    'Program',
    'Level',
    'Duration_Years',
    'Exchange_Rate',
    'total_cost_index'  # Using the newly created feature
]

# Select only columns that exist in the dataset
df_selected = df[[
    col for col in columns_of_interest if col in df.columns
]].copy()

print("\n=== Selected Features for Analysis ===")
print(f"Features: {df_selected.columns.tolist()}")
print(f"\nFirst 5 rows:")
df_selected.head()
```

```python
"""
TASK 3: SQLite Database Operations

Purpose:
    Demonstrate database creation and data management using SQLite.

Operations:
    1. Create/connect to SQLite database (sample.db)
    2. Load CSV files from sql/ directory into database tables
    3. Create and populate an 'employees' table with sample data
    4. Query and display data from the database

Note:
    This task demonstrates database integration capabilities.
    Database is created with proper schema and data validation.
"""

import pandas as pd
import sqlite3
import os

# Step 1: Create/Connect to SQLite Database
db_path = 'sample.db'
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

print(f"Connected to SQLite database: {db_path}")

# Step 2: Load CSV Files from sql/ Directory
csv_files = [
    'sql/brands.csv',
    'sql/categories.csv',
    'sql/customers.csv',
    'sql/order_items.csv',
    'sql/orders.csv',
    'sql/products.csv',
    'sql/staffs.csv',
    'sql/stocks.csv',
    'sql/stores.csv'
]

print("\n=== Loading CSV Files into Database ===")
for file_path in csv_files:
    if os.path.exists(file_path):
        table_name = os.path.basename(file_path).replace('.csv', '')
        df_temp = pd.read_csv(file_path)
        df_temp.to_sql(table_name, conn, if_exists='replace', index=False)
        print(f"✓ Table '{table_name}' created ({len(df_temp)} rows)")
    else:
        print(f"✗ File not found: {file_path}")

# Step 3: Create Employees Table with Sample Data
print("\n=== Creating Employees Table ===")
cursor.execute('''
    CREATE TABLE IF NOT EXISTS employees (
        id INTEGER PRIMARY KEY,
        name TEXT NOT NULL,
        department TEXT NOT NULL,
        salary REAL NOT NULL
    )
''')
print("✓ Employees table created")

# Step 4: Insert Sample Records
employees_data = [
    (1, 'Alice Johnson', 'HR', 65000),
    (2, 'Bob Smith', 'Engineering', 92000),
    (3, 'Charlie Davis', 'Marketing', 58000),
    (4, 'David Brown', 'Sales', 62000),
    (5, 'Eve Wilson', 'Engineering', 95000)
]

cursor.executemany(
    'INSERT OR REPLACE INTO employees VALUES (?, ?, ?, ?)',
    employees_data
)
conn.commit()
print(f"✓ Inserted {len(employees_data)} employee records")
```

```python
"""
Retrieve and Display Data from SQLite Database

Purpose:
    Query the database and display the created employees table.
    Demonstrates reading data from SQLite into pandas DataFrames.

Output:
    Display of the employees table with all records and statistics.
"""

print("\n=== Querying Employees Table ===")
df_employees = pd.read_sql_query("SELECT * FROM employees", conn)

print("\nEmployees Data:")
print(df_employees)

print("\nStatistics:")
print(f"Total Employees: {len(df_employees)}")
print(f"Average Salary: ${df_employees['salary'].mean():,.2f}")
print(f"Salary Range: ${df_employees['salary'].min():,.2f} - ${df_employees['salary'].max():,.2f}")

# Close database connection
conn.close()
print("\n✓ Database connection closed")
```

```python
"""
TASK 1: Fetch Data from TMDB API

Purpose:
    Retrieve movie data from The Movie Database (TMDB) API.
    Demonstrates API integration and data transformation.

Process:
    1. Define API endpoint and authentication key
    2. Issue HTTP request to fetch popular movies
    3. Extract relevant fields from JSON response
    4. Convert to pandas DataFrame and save to CSV

Output:
    tmdb_movies.csv - Contains title, release_date, rating, popularity, language

Note:
    API Key should be kept secure. For production, use environment variables.
"""

import requests
import pandas as pd

# Step 1: Define API Configuration
API_KEY = "ba350f8d4b309caa593ad9debab3c11a"  # Replace with your actual key
API_ENDPOINT = "https://api.themoviedb.org/3/movie/popular"

print("TMDB API - Popular Movies Fetcher")
print(f"Endpoint: {API_ENDPOINT}")

# Step 2: Fetch Data from API
try:
    response = requests.get(
        API_ENDPOINT,
        params={"api_key": API_KEY}
    )
    response.raise_for_status()  # Raise exception for bad status codes
    data = response.json()
    print(f"✓ API request successful (Status: {response.status_code})")
except requests.exceptions.RequestException as e:
    print(f"✗ API request failed: {e}")
    data = {"results": []}

# Step 3: Extract and Transform Data
movies_list = []
for movie in data.get('results', []):
    movies_list.append({
        "title": movie.get('title', 'N/A'),
        "release_date": movie.get('release_date', 'N/A'),
        "rating": movie.get('vote_average', 0),
        "popularity": movie.get('popularity', 0),
        "language": movie.get('original_language', 'N/A')
    })

print(f"\n✓ Extracted {len(movies_list)} movies")

# Step 4: Create DataFrame and Save to CSV
df_tmdb_movies = pd.DataFrame(movies_list)
df_tmdb_movies.to_csv('tmdb_movies.csv', index=False)

print(f"✓ Saved to tmdb_movies.csv")
print("\n=== Top 5 Popular Movies ===")
print(df_tmdb_movies.head(5))
```
